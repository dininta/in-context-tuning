```python
import matplotlib.pyplot as plt
import re

with open('output/log_train_classification_test_classification.txt') as fin:
    lines = fin.readlines()

train_losses = []
val_losses = []
for line in lines:
    match = re.search(r'Train loss: (.*?)\n', line)
    if match:
        train_losses.append(float(match.group(1)))
        continue

    match = re.search(r'Validation loss: (.*?)\n', line)
    if match:
        val_losses.append(float(match.group(1)))

plt.plot(train_losses, label='train')
plt.plot(val_losses, label='val')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()
plt.show()
```
